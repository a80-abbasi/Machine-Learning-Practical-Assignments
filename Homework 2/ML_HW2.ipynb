```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

r1 = np.random.randn(50)/3
r2 = np.random.randn(50)/3
scatter_hue = np.random.choice([True, False], size=50)
r1 += np.ones(50) * ~scatter_hue * 0.5
axes[0][0].set_title('Scatter plot')
sns.scatterplot(x=r1, y=r2, hue=scatter_hue, ax=axes[0][0])

dist_smaples = list(np.random.normal(loc=0.0, scale=1, size=20)) + [8,8,-32]
axes[0][1].set_title('Histogram')
sns.distplot(dist_smaples, bins=20, hist=True, ax=axes[0][1])

pie_values = [10, 30, 2, 45]
axes[1][0].set_title('Pie Chart')
axes[1][0].pie(sorted(pie_values), labels=sorted(pie_values))

box_values = np.concatenate((2 + np.random.randn(20), -5 + 2 * np.random.randn(5)))
axes[1][1].set_title('Box Plot')
sns.boxplot(box_values, ax=axes[1][1]);
```

        برای ستون‌های عددی نیز برخی مقادیر غیر مجاز یا ناموجود هستند(یعنی برای ستون‌هایی که باید مقدار مثبت داشته باشند، مقدار منفی ثبت شده است و یا داده‌ای بسیار بزرگ به عبارتی outlier وجود دارد) که باید به صورت None در بیایند.
    این ستون‌ها را تغییر دهید و دوباره نسبت مقادیر ناموجود هر ستون و همینطور اطلاعات آماری ستون های عددی تغییر یافته را محاسبه کنید.

```python
# check datatypes of values:
# reports, dependents, majorcards, active should contain integer values:
bank.dtypes
```

```python
# create a dataframe for quantiles and whisks for numerical columns:
bank_numerical_stat = bank_numerical_stat.T
bank_numerical_stat['iqr'] = bank_numerical_stat['75%'] - bank_numerical_stat['25%']
bank_numerical_stat['whisk2'] = bank_numerical_stat['75%'] + bank_numerical_stat['iqr'] * 1.5
bank_numerical_stat['whisk1'] = bank_numerical_stat['25%'] - bank_numerical_stat['iqr'] * 1.5
bank_numerical_stat
```

```python
# apply changes named above:

categorical_cols = bank_categorial.columns.tolist()
numeric_cols = bank_numerical.columns.tolist()
discrete_cols = ['majorcards', 'active']

# change categorial values to boolean and change invalid values to NaN
bank[categorical_cols] = bank[categorical_cols].replace(['yes', 'no'], [True, False]).applymap((lambda x: x if isinstance(x, bool) else np.nan))

# numeric columns can't have negative valeus
bank[numeric_cols] = bank[numeric_cols].applymap(lambda x: np.nan if x < 0 else x)

# discrete columns should have discrete values:
bank[discrete_cols] = bank[discrete_cols].applymap(lambda x: int(x) if x.is_integer() else np.nan)

# delete outliers with respect to whiskers:
# age shouldn't be below lower whisker
bank['age'] = bank['age'].map(lambda x: np.nan if x < bank_numerical_stat.loc['age', 'whisk1'] else x)
# share shouldn't be above upper whisker
bank['share'] = bank['share'].map(lambda x: np.nan if x > bank_numerical_stat.loc['share', 'whisk2'] else x)
bank
```

<div dir='rtl'>پاسخ:
<br>
در روش PCA، ما ابتدا داده‌ها را منهای میانگین آن‌ها می‌کنیم تا میانگین داده‌های حاصل صفر باشد. زیرا باید داده‌های ما zero-based باشند و در این صورت PCA می‌توانند جهتی را پیدا کنند که واریانس داده‌ها بیشترین حالت ممکن باشد و میانگین داده‌ها را به عنوان متغیر مهمی در نظر نمی‌گیرد. و این مرحله بخش اجباری PCA است. <br>
تقسیم کردن داده‌ها بر انحراف معیار آن‌ها اجباری نمی‌باشد. اما در اینجا ما مقادیر ستون‌ها را بر انحراف معیار آن‌ها هم تقسیم کردیم. و دلیل این کار این است که مقادیر ستون‌های دیتافریم ما از جنس‌های مختلف و در اندازه‌های مختلف هستند و اگر این تقسیم را انجام ندهیم، ستون‌هایی که مقادیر بزرگ‌تری دارند، تاثیر بیشتری در مقدار واریانس داده‌ها خواهند داشت و PCA به روشی transform داده‌ها به فضای با بعد پایین را انجام می‌دهد که به این مقادیر بزرگ‌تر اهمیت بیشتری داده است تا واریانسشان را تا جای ممکن کم کند. <br>
بنابراین ما برای این که کاری کنیم که همه‌ی ستون‌ها تاثیر مشابهی در واریانس داده‌های حاصل داشته باشند، هر ستون را تقسیم بر انحراف معیار نیز می‌کنیم تا واریانس داده‌های حاصل یک شود.

</div>

```python
dataset = bank.copy()
dataset = (dataset - dataset.mean()) / np.std(dataset, axis=0)
```

<div dir=rtl>
حال ماتریس کوواریانس دیتاست موجود را بدست آورید و مقادیر و بردار‌های ویژه‌ی آن را به ترتیب نزولی مقادیر ویژه نمایش دهید.
    <br>
    (۳ نمره)
</div>

```python
cov_matrix = np.divide(np.dot(dataset.T, dataset), len(dataset))  # X^t * X / N
eigenvalues, eigenvectors, = np.linalg.eig(cov_matrix)
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]
print(f'eigenvalues:\n{eigenvalues}\n')
print(f'eigenvectors:\n{eigenvectors}\n')
```

<div dir=rtl>
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
    <br>
    (۴ نمره)
</div>

```python
variance_ratio = eigenvalues / sum(eigenvalues) * 100
labels = [f'ev{i}' for i in range(1, len(variance_ratio) + 1)]
fig1, ax1 = plt.subplots(figsize=(8, 8))
ax1.pie(variance_ratio, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

<div dir=rtl>
با حفظ حداقل ۵۰ درصد واریانس داده‌ها، ماتریس پرتو(projection matrix) را بدست آورید و تبدیل pca را بر روی دیتاست اجرا کنید.
    <br>
    (۵ نمره)
</div>

```python
csum = np.cumsum(eigenvalues)
pc_num = bisect(csum, csum[len(csum) - 1] / 2) + 1
projection_matrix = eigenvectors[0:pc_num].T
projected_data = np.dot(dataset, projection_matrix)
projected_data
```

    <br>
    (۴ نمره)

```python
fig = plt.figure(figsize=(10, 12))
sns.scatterplot(x=projected_data.T[0], y=projected_data.T[1], hue=wealthy)
```

همانطور که می‌بینیم و پیش‌بینی می‌شد، چون PCA مبنای کارش در جدا کردن داده‌ها بر اساس لیبل‌ها نیست و تنها سعی می‌کند جهتی را بیابد که افکنش داده‌ها در آن بیش‌ترین واریانس را داشته باشند، دو دسته‌ی فقیر و ثروتمند از هم کاملا جدا نمی‌شوند. <br>
اما با این حال، همانطور که مشاهده می‌شود فقیر یا ثروتمند بودن (درواقع income آن‌ها. چون ستون wealth در PCA نیست) در جدا کردن نقاط از یکدیگر تاثیر داشته است و می‌توان به طور تقریبی خطی رسم کرد که اکثر افراد فقیر در یک سمت آن، و اکثر افراد ثروتمند در سمت دیگر آن قرار دارند.
